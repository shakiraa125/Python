```python
plt.plot(df['date'],df['price'],'.')
plt.title('date & price')
plt.xlabel('date') #label for x axis
plt.ylabel('price')
plt.show()
```

```python
plt.plot(df['condition'],df['price'],'.')
plt.title('date & price')
plt.xlabel('date') #label for x axis
plt.ylabel('price')
plt.show()
```

```python
avg=df.groupby('country')['price'].mean()
print(avg)
```

```python
avg.plot(kind='bar')
plt.xlabel('country')
plt.ylabel('price')
plt.title('average price by country')
plt.show()
```

```python
plt.hist([df['bathrooms'],df['bedrooms']],label=['bathroom','bedroom'])
plt.xlabel('bathroom')
plt.ylabel('bedroom')
plt.title('distribution the bathroom and bedroom')
plt.legend()
plt.show()
```

```python
avg=df.groupby('floors')['sqft_above'].mean()
print(avg)
avg.plot(kind='bar',color=['grey','pink'])#kind is for which graph 
plt.xlabel('floors')
plt.ylabel('avg sqft_above ')
plt.title('avg levels by sqft_above')
```

```python
custom_bins=[0,10,20,30,40,50,60]
sns.histplot(data=df,x="price",bins=custom_bins)
plt.show()
```

```python
sns.histplot(data=df,x='price',kde=True,stat='probability')
```

```python
sns.scatterplot(data=df,x="price",y="bathrooms")
plt.show()
```

```python
sns.barplot(x="floors",y="price",data=df)
plt.xlabel("price")
plt.ylabel("floors")
plt.title("Relationship between date floors and price") 
plt.show()
```

```python
sns.barplot(x="floors",y="price",data=df,ci=None)
plt.xlabel("floors")
plt.ylabel("price")
plt.title("Relationship between total price and floors") 
plt.show()
```

```python
sns.barplot(x="floors",y="price",data=df,ci=None,palette='pastel')
plt.xlabel("floors")
plt.ylabel("price")
plt.title("Relationship between total price and floors") 
plt.show()
```
